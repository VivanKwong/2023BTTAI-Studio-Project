```python
# since the 'fax' feature has no data and the 'url' feature contain mostly missing data, we can remove them
rest_geo_df.drop(columns=['fax','url'], inplace=True)
rest_geo_df.shape
```

```python
# check the values in the 'address' feature
value_counts = rest_geo_df['address'].value_counts()
value_counts
```

```python
# note that there are 27 entries with missing value
# since the feature should be unique, we can't replace them with the mode
# replace them with unique placeholders

missing_indices = rest_geo_df[rest_geo_df['address'] == '?'].index
for i, idx in enumerate(missing_indices, start=1):
    rest_geo_df.at[idx, 'address'] = f'?{i}'

value_counts = rest_geo_df['address'].value_counts()
value_counts
```

```python
# find missing values and replace them with mean/mode
# a function to identify numeric/categorical data
def is_numeric(series):
    try:
        pd.to_numeric(series, errors='raise')
        return True
    except:
        return False

# Function to get the second mode, given a column
def get_second_mode(column):
    value_counts = column.value_counts()
    if value_counts.index[0] == '?':
        return value_counts.index[1]
    else:
        return value_counts.index[0]

# Iterate through each column in the DataFrame
for col in rest_geo_df.columns:
    # If the column is numeric
    if is_numeric(rest_geo_df[col]):
        # Convert ? to NaN
        rest_geo_df[col] = pd.to_numeric(rest_geo_df[col], errors='coerce')
        # Fill NaN with mean
        rest_geo_df[col].fillna(rest_geo_df[col].mean(), inplace=True)
    else:
        # For categorical columns, replace ? with the most frequent value
        mode_value = get_second_mode(rest_geo_df[col])
        rest_geo_df[col] = rest_geo_df[col].replace('?', mode_value)
```

```python
# note that they have capitalized vs uncapitalized issues
# convert all of the entries in the 'city','state',and'country' features to lowercases
for col in columns_capitalized:
  rest_geo_df[col] = rest_geo_df[col].str.lower()

# handle the 'slp' entry and the 'san luis potosi' entry
variations1 = ['s.l.p.','slp','s.l.p']
variations2 = ['san luis potosi','san luis potos','san luis potosi ']
variations3 = ['cd victoria','ciudad victoria','cd. victoria']
variations4 = ['victoria','victoria ']
for col in columns_capitalized:
  rest_geo_df[col] = rest_geo_df[col].replace(variations1, 's.l.p')
  rest_geo_df[col] = rest_geo_df[col].replace(variations2, 'san luis potosi')
  rest_geo_df[col] = rest_geo_df[col].replace(variations3, 'cd victoria')
  rest_geo_df[col] = rest_geo_df[col].replace(variations4, 'victoria')

# view the dataset after processing
for col in columns_capitalized:
  print(rest_geo_df[col].value_counts(),'\n')
```

**Matrix Factorization with User Ratings**

The "UsersRestaurantRatings" file has been altered in the following ways in "UsersRestaurantRatingsNew":
1. A new column, "userIDnum" has been created, which strips the excess characters from the "userID" column to isolate the user ID as an integer.
2. The "rating," "food_rating," and "service_rating" columns' values have all been increased by 1. Now, instead of the user ratings being either 0, 1, or 2 for each restaurant, they will be 1, 2, or 3. This will help when it comes to creating the user-restaurant ratings matrix.
3. The data has been sorted, first in order of increasing userID and then in order of increasing placeID.

```python
# there are 138 unique users and 130 unique restaurants in this dataset
print(user_ratings_df['userID'].nunique())
print(user_ratings_df['placeID'].nunique())
```

```python
# so our matrix where each user has a row and each restaurant has a column
# will have 17940 total elements
138*130
```

```python
# put the data into a matrix form, where any restaurants users didn't rate will get a 0
# start by establishing a list of 17940 zeroes
ratings = 0/np.arange(start=1, stop=17941, step=1)
# k will track which element in the matrix out of 17940 we're on
k = 0
# i will track which entry out of the 1161 in the data we're on
i = 0
# restaurants has a sorted list of the 130 unique restaurant IDs
restaurants = np.unique(user_ratings_df['placeID'])
# go through each user's row in the matrix
for user in range(1, 139):
  # go through each restaurant for that user
  for rest in restaurants:
    # check to see whether that user has rated that restaurant
    if (user_ratings_df.loc[i]['placeID']==rest and user_ratings_df.loc[i]['userIDnum']==user and i<1160):
      # if that user has rated that restaurant, change the 0 in that spot to their rating
      ratings[k] = user_ratings_df.loc[i]['rating']
      # increment i by one to move to the next user rating in the data
      i = i+1
    # increment k by one to move to the next element in the ratings list
    k = k+1
```

```python
# we're missing the last rating from the last user
sum(ratings!=0)
```

```python
# manually change that one so all 1161 ratings are correct in the matrix
ratings[17863] = 3.0
sum(ratings!=0)
```

```python
# make the list of ratings into a matrix with 138 rows and 130 columns
ratings = np.array(ratings)
ratings.shape = (138,130)
print(ratings)
```

```python
# calculate each user's average rating, then plot the spread
userAverages = list()
for user in range(1,139):
  userAverages.append(sum(ratings[user-1,0:129])/sum(ratings[user-1,0:129]!=0))
fig, ax = plt.subplots(figsize=(4,4))
sns.histplot(data=userAverages)
ax.set_xlabel("Average User Ratings")
ax.set_ylabel("Number of Users")
plt.show()
```

```python
# calculate each restaurant's average rating, then plot the spread
restAverages = list()
for rest in range(0,130):
  restAverages.append(sum(ratings[0:137,rest])/sum(ratings[0:137,rest]!=0))
fig, ax = plt.subplots(figsize=(4,4))
sns.histplot(data=restAverages)
ax.set_xlabel("Average Restaurant Ratings")
ax.set_ylabel("Number of Restaurants")
plt.show()
```

```python
# get the eigenvalues and eigenvectors from the ratings matrix
import math
from numpy.linalg import eig
eigenvals, rtreigenvects = eig(ratings.T@ratings)
```

```python
# sort the eigenvalues from biggest to smallest
eigenvals = np.sort(eigenvals)[::-1]
# we need the first 51 eigenvalues to get 90% of their total sum
eigenvalsum = sum(eigenvals)
sum(eigenvals[0:51])/eigenvalsum
```

```python
# use this function for the decomposition
from numpy.linalg import svd
U, d, V = svd(ratings)
print(U.shape)
print(d.shape)
print(V.shape)
```

```python
# only keep the first 51 columns for U
U = U[0:138,0:51]
# only keep the first 51 eigenvalues for D
d = d[0:51]
# only keep the first 51 rows for V
V = V[0:51,0:130]
print(U.shape)
print(d.shape)
print(V.shape)
```

```python
# make d into a matrix where the eigenvalues are on the diagonal
d = np.diag(d)
print(d.shape)
```

```python
# multiply U, d, and V together to get a 138x130 matrix approximating the original
ratingsPredicted = U@d@V
print(ratingsPredicted.shape)
```

```python
# evaluate how closely the predictions match the actual ratings
import sklearn.metrics
rmse = math.sqrt(sklearn.metrics.mean_squared_error(ratings, ratingsPredicted))
print(rmse)
```

```python
# find out which restaurant that each user hasn't been to yet was predicted to have the highest rating
maxRatings = [] # store list of highest predicted ratings for each user
maxIndices = [] # store list of recommended restaurants for each user
# loop through each user
for user in range(0,138):
  maxRating = 0 # store the user's highest predicted rating for a restaurant
  maxIndex = 0 # store the index of the user's highest predicted restaurant
  # loop through each restaurant for that user
  for rest in range(0,130):
    # if the predicted rating is high and the user hasn't rated that restaurant
    if(ratingsPredicted[user,rest]>maxRating and ratings[user,rest]==0):
      # store the rating of that restaurant and which restaurant it is
      maxRating = ratingsPredicted[user,rest]
      maxIndex = rest
  # add that user's maximum predicted rating and recommended restaurant to lists
  maxRatings.append(maxRating)
  maxIndices.append(maxIndex)
# print the list of restaurants that were recommended to each user
print(maxIndices)
```

```python
# randomly select 232 ratings to remove
np.random.seed(12)
ratingsToRemove = np.random.choice(user_ratings_df.index,232,replace=False)
ratingsToRemove = np.sort(ratingsToRemove)
print(ratingsToRemove)
```

```python
# keep track of which ratings we removed from which users and restaurants
usersRemoved = []
restsRemoved = []
ratingsRemoved = []
for ratingToRemove in ratingsToRemove:
  row = (user_ratings_df.loc[ratingToRemove][1])-1
  usersRemoved.append(row)
  col = list(restaurants).index(user_ratings_df.loc[ratingToRemove][2])
  restsRemoved.append(col)
  ratingsCopy[row][col] = 0
  ratingsRemoved.append(ratings[row][col])
print(len(ratingsRemoved))
```

```python
# make sure that we only removed non-zero ratings
np.sum(ratingsCopy!=0)
```

```python
# re-run SVD with the 929 remaining ratings
svd(ratingsCopy)
U_new, d_new, V_new = svd(ratingsCopy)
print(U_new.shape)
print(d_new.shape)
print(V_new.shape)
```

```python
# only keep the first 51 columns for U
U_new = U_new[0:138,0:51]
# only keep the first 51 eigenvalues for D
d_new = d_new[0:51]
# only keep the first 51 rows for V
V_new = V_new[0:51,0:130]
print(U_new.shape)
print(d_new.shape)
print(V_new.shape)
```

```python
# make d into a matrix where the eigenvalues are on the diagonal
d_new = np.diag(d_new)
print(d_new.shape)
```

```python
# multiply U, d, and V together to get a 138x130 matrix approximating the original
ratingsPredictedNew = U_new@d_new@V_new
print(ratingsPredictedNew.shape)
```

```python
# evaluate how closely the predictions match the actual ratings
rmseNew = math.sqrt(sklearn.metrics.mean_squared_error(ratingsCopy, ratingsPredictedNew))
print(rmseNew)
```
